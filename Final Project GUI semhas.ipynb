```python
import numpy
from numpy import *
import cv2
import PIL
import datetime
import time
import csv
## masukkan model
from keras.models import model_from_json
json_file=open("cnn98k4.json", "r")
loaded_model_json=json_file.read()
json_file.close()
loaded_model=model_from_json(loaded_model_json)
loaded_model.load_weights("cnn98k4.h5")
loaded_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

from tkinter import ttk
from tkinter import *
from tkinter.messagebox import showinfo

def identity(X):
    if loaded_model.predict_classes(X)[0].astype('str') == '0':
        f = "dev.PNG"
    elif loaded_model.predict_classes(X)[0].astype('str') == '1':
        f = "san.PNG"
    elif loaded_model.predict_classes(X)[0].astype('str') == '2':
        f = "ulf.PNG"
    elif loaded_model.predict_classes(X)[0].astype('str') == '3':
        f = "vin.PNG"
    return f

def identity2(X):
    if loaded_model.predict_classes(X)[0].astype('str') == '0':
        g = '06211540000115'
    elif loaded_model.predict_classes(X)[0].astype('str') == '1':
        g = '06211540000123'
    elif loaded_model.predict_classes(X)[0].astype('str') == '2':
        g = '06211540000021'
    elif loaded_model.predict_classes(X)[0].astype('str') == '3':
        g = '06211540000131'
    return g

def employee():
    ident=Toplevel()
    ident.title("Employee Information")
    imdev = PhotoImage(file="dev.PNG")
    dev = Label(ident, image=imdev)
    dev.pack()
    imsan=PhotoImage(file="san.png")
    san = Label(ident, image=imsan)
    san.pack()
    imulf=PhotoImage(file='ulf.png')
    ulf = Label(ident, image=imulf)
    ulf.pack()
    imvin=PhotoImage(file="vin.PNG")
    vin = Label(ident, image=imvin)
    vin.pack()
    ident.mainloop()

def about():
    top=Toplevel()
    top.title("About")
    text="""
This graphic user interface made by:
Ulfa Siti Nuraini
using for "Face Recognition"

for more information you can send message to:
ulfanuraini24@gmail.com

June, 2019
"""
    text2 = Text(top, height=15, width=60)
    text2.insert(END, text)
    text2.pack()

def company():
    top=Toplevel()
    top.title("Company")
    label = Label(top, text="VIVACIOUS 2015")
    label.config(font=('Arial', 25, 'bold'))
    label.pack()
    logo=PhotoImage(file='logo.png')
    lala = Label(top, image=logo)
    lala.pack()
    top.mainloop()
    
def writeToFile(y):
    now = datetime.datetime.now()
    with open('Book1.csv', 'a') as f:
        w = csv.writer(f, lineterminator='\n')
        if y=='06211540000115':
            x='Devita Prima Vernanda'
        elif y=='06211540000123':
            x='Mochamad Ihsan Ananto'
        elif y=='06211540000021':
            x='Ulfa Siti Nuraini'
        elif y=='06211540000131':
            x='Mohammad Haidar Alvin Purwana'
        w.writerow([now.strftime("%Y-%m-%d %H:%M:%S"),y,x])
    return f

def camera():
    img_counter = 0
    webcam = cv2.VideoCapture(0) #Use camera 0
    while(True):
        (rval, im) = webcam.read()
        im=cv2.flip(im,1,0)
        cv2.imshow('Press space to capture!',im)
        key = cv2.waitKey(10) 
        if key == 27:
            break
        elif key == 32:
            img_name = "image1.png".format(img_counter)
            minisize = (360,270)
            mini = cv2.resize(im, minisize)
            cv2.imwrite(img_name,mini)
            cv2.imwrite("image1.JPG",im)
            print("[] written".format(img_name))
            img_counter +=1
            webcam.release()
            cv2.destroyAllWindows()           
            
            photo = PhotoImage(file='image1.png')
            text1.delete(1.0, END)
            text1.image_create(END,image=photo)
            
            from PIL import Image
            im = Image.open('image1.jpg')
            img = im.convert('L')
            (newW, newH)=(160,120)
            imgnew=PIL.Image.new("L", (newW, newH), "white")
            for col in range(newW):
                for row in range(newH):
                    p=img.load()
                    z=p[col/0.25,row/0.25]
                    q=imgnew.load()
                    q[col,row]=z
            X = []
            X = array([array(imgnew).flatten()],'f')
            X = X.reshape(X.shape[0], 1, 120, 160)
            X = X.astype('float64')
            X /= 255
            print(loaded_model.predict_classes(X))
            f=identity(X)
            imclass = PhotoImage(file=f)
            text4.delete(1.0, END)
            text4.image_create(END,image=imclass)
    cv2.destroyAllWindows()

def database():
    from PIL import Image
    im = Image.open('image1.jpg')
    img = im.convert('L')
    (newW, newH)=(160,120)
    imgnew=PIL.Image.new("L", (newW, newH), "white")
    for col in range(newW):
        for row in range(newH):
            p=img.load()
            z=p[col/0.25,row/0.25]
            q=imgnew.load()
            q[col,row]=z
    X = array([array(imgnew).flatten()],'f')
    X = X.reshape(X.shape[0], 1, 120, 160)
    X = X.astype('float64')
    X /= 255
    print(loaded_model.predict_classes(X))
    g = identity2(X)
    yuhuu = writeToFile(g)
    popup_showinfo()

def popup_showinfo():
    showinfo("window", "Saving done!")

def tick():
    time2=time.strftime('%H:%M:%S')
    clock.config(text=time2)
    clock.after(200,tick)
            
root = Tk()
root.title("Graphic User Interface for Face Recognition")
root.configure(background='gray')

clock = Label(root, font=('times',20,'bold'), bg='gray')
clock.pack(fill=BOTH,expand=1)
tick()

menubar=Menu(root)
filemenu=Menu(menubar, tearoff=0)
filemenu.add_command(label="Company", command=company)
filemenu.add_command(label="Employee", command=employee)
menubar.add_cascade(label="About Us", menu=filemenu)
menubar.add_command(label="About",command = about)
menubar.add_command(label="Quit",command = root.destroy)
    
root.config(menu=menubar)

btn1 = Button(root, text = "Open Camera")
btn1.config(command=camera)
btn1.config(bg='white', height=2, width=51, font=('Consolas', 10))
btn1.pack()

text1 = Text(root, height=15, width=45)
photo = PhotoImage(file='Untitled.png')
text1.pack()

Frame(height=2, bd=1, relief=SUNKEN).pack(fill=X, padx=1,pady=10)
text4 = Text(root, height=12, width=75)
pt = PhotoImage(file='Blank.png')
text4.image_create(END,image=pt)
text4.pack()

btn2 = Button(root, text = "Save to Database", command=database)
btn2.config(bg='white', height=2, width=85, font=('Consolas', 10))
btn2.pack()

root.mainloop()
```
